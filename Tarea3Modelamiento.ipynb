```python
from random import randint
import math

lambd = float(input("Ingrese un valor de lambda: "))
mu = float(input("Ingrese un valor de mu: "))

if lambd < mu:
    num_servidores = list(range(2, 11))
    k = num_servidores.pop(0)  # Siempre elimina el primer número de la lista (2)
    lambd *= k  # Multiplicar lambda por el número de servidores
    print("El valor final de lambda después de la multiplicación es:", lambd)
else:
    k = int(input("Ingrese el número de servidores: "))


def probabilidad_sistema_vacio(K, lambd, mu):
  sum_term = sum((lambd / mu)**n / math.factorial(n) for n in range(K))
  additional_term = (k * mu * (lambd / mu)**K / (math.factorial(K) * (K * mu - lambd)))
  return (sum_term + additional_term)**-1

def numero_promedio_entidades_cola(P0, lambd, mu, K):
    bracket_term = P0 * (lambd * mu * (lambd / mu)**K / (math.factorial(K - 1)) / ((K * mu - lambd)**2))
    L_q = bracket_term
    return L_q

def numero_promedio_entidades_sistema(Lq, lambd, mu):
    return Lq + (lambd / mu)

def tiempo_promedio_espera_entidad_cola(Lq, lambd):
     return Lq / lambd

def tiempo_promedio_espera_entidad_sistema(Wq, mu):
    return Wq + 1 / mu

def probabilidad_sistema_ocupado(P0, K, lambd, mu):
    prob_n_geq_K = P0 * ((K * mu * (lambd / mu)**K) / (math.factorial(K) * (K * mu - lambd)))
    return prob_n_geq_K

def probabilidad_n_unidades_sistema(n, K, lambd, mu, P0):
    if n < K:
       return P0 * ((lambd / mu)**n / math.factorial(n))
    else:
       return P0 * ((lambd / mu)**n / (math.factorial(K) * K**(n - K)))


P0 = probabilidad_sistema_vacio(k, lambd, mu)
Lq = numero_promedio_entidades_cola(P0, lambd, mu, k)
L = numero_promedio_entidades_sistema(Lq, lambd, mu)
Wq = tiempo_promedio_espera_entidad_cola(Lq, lambd)
W = tiempo_promedio_espera_entidad_sistema(Wq, mu)
Po = probabilidad_sistema_ocupado(P0, k, lambd, mu)
Pn = probabilidad_n_unidades_sistema(5, k, lambd, mu, P0)

print("La probabilidad del sistema vacío:", P0)
print("La probabilidad del número promedio de entidades en la cola:", Lq)
print("La probabilidad del número promedio de entidades del sistema:", L)
print("La probabilidad del tiempo promedio de espera de la entidad en cola:", Wq)
print("La probabilidad del tiempo promedio de espera de la entidad2 del sistema:", W)
print("La probabilidad del sistema ocupado:", Po)
print("La probabilidad de n unidades del sistema:", Pn)
```

```python
import math

def modelo_mm1(lambd, mu):
    rho = lambd / mu
    L = rho / (1 - rho)
    Lq = (lambd**2) / (mu * (mu - lambd))
    W = 1 / (mu - lambd)
    Wq = lambd / (mu * (mu - lambd))
    P0 = 1 - rho
    return P0, Lq, L, Wq, W, Po

def modelo_mmk(lambd, mu, k):
    rho = lambd / (k * mu)
    P0 = 1 / (sum([(lambd/mu)**n / math.factorial(n) for n in range(k)]) +
               ((lambd/mu)**k / (math.factorial(k) * (1 - rho))))
    Lq = P0 * ((lambd/mu)*k * rho) / (math.factorial(k) * (1 - rho)*2)
    L = Lq + lambd / mu
    Wq = Lq / lambd
    W = Wq + 1 / mu
    return P0, Lq, L, Wq, W, Po

def modelo_mg1(lambd, mu, sigma):
    rho = lambd / mu
    P0 = 1 - rho
    Lq = (lambd*2 * sigma**2 + rho*2) / (2 * (1 - rho))
    L = Lq + rho
    Wq = Lq / lambd
    W = Wq + 1 / mu
    Pw = rho
    return P0, Lq, L, Wq, W, Pw

def modelo_md1(lambd, mu):
    rho = lambd / mu
    P0 = 1 - rho
    Lq = (rho**2) / (2 * (1 - rho))
    L = Lq + rho
    Wq = lambd / (2 * mu * (mu - lambd))
    W = lambd * (2 * mu - lambd) / (2 * mu * (mu - lambd))
    Pw = rho
    return P0, Lq, L, Wq, W, Pw

def modelo_mm1_q(lambd, mu, q):
    rho = lambd / mu
    P0 = (1 - rho) / (1 - rho**(q+1))
    Pq = (rho*q * (1 - rho)) / (1 - rho*(q+1))
    L = (rho / (1 - rho)) - ((q + 1) * rho*(q+1)) / (1 - rho*(q+1))
    Lq = L - (1 - P0)
    Wq = Lq / (lambd * (1 - Pq))
    return P0, Lq, L, Wq, W, Po

def modelo_mm1_n(lambd, mu, n):
    rho = lambd / mu
    P0 = 1 / sum([(n / (n - i)) * (lambd / mu)*i if i < n else (lambd / mu)*i for i in range(n+1)])
    L = n - (mu / lambd) * (1 - P0)
    Lq = L - (1 - P0)
    Wq = Lq / ((n - L) * lambd)
    W = Wq + 1 / mu
    return L, Lq, W, Wq, P0

def modelo_mmk_n(lambd, mu, k, N):
    # Factor de utilización
    U = lambd / (k * mu)

    # Probabilidad de que el sistema esté vacío (P0)
    sumatoria1 = sum([(lambd/mu)**n / math.factorial(n) for n in range(k)])
    sumatoria2 = sum([(lambd/mu)**n / (math.factorial(N - n) * math.factorial(k) * (n - k) * k**(n - k)) / ((lambd/mu)**(n - 1)) if n != 0 else 0 for n in range(k, N + 1)])
    P0 = 1 / (sumatoria1 + sumatoria2)

    # Número promedio de entidades en la cola (Lq)
    Lq = sum([(n - k) * P0 * (lambd/mu)**n / (math.factorial(N - n) * math.factorial(k) * (n - k)) for n in range(k, N + 1)])

    # Número promedio de entidades en el sistema (L)
    L = Lq + k

    # Tiempo promedio de una entidad en la cola (Wq)
    Wq = Lq / lambd if lambd != mu * k else 0

    # Tiempo promedio de una entidad en el sistema (W)
    W = Wq + 1 / mu

    # Probabilidad de que el sistema esté ocupado
    Pw = 1 - P0

    return P0, Lq, L, Wq, W, Pw

def analisis_costo(lambd, mu, Cs, Cw):
    k_opt = 1
    CT_min = float('inf')

    for k in range(1, 100):
        L, _, _, _, _ = modelo_mmk(lambd, mu, k)
        CT = Cs * k + Cw * L
        if CT < CT_min:
            CT_min = CT
            k_opt = k
        else:
            break

    return k_opt, CT_min

def calcular_probabilidad_n(lambd, mu, k, n):
    L, Lq, _, _,_,P0 = modelo_mmk(lambd, mu, k)
    probabilidad_n = P0 * math.comb(n, k) * (lambd/mu)**n
    return probabilidad_n

lambd = float(input("Ingrese un valor de lambda: "))
mu = float(input("Ingrese un valor de mu: "))

metodo = input("Ingrese el método a utilizar (mm1, mmk, mg1, md1, mm1_q, mm1_n, mmk_n): ")

if metodo == "mmk" or metodo == "mmk_n":
    if lambd < mu:
        num_servidores = list(range(2, 11))
        k = num_servidores.pop(0)  # Siempre elimina el primer número de la lista (2)
        lambd *= k  # Multiplicar lambda por el número de servidores
        print("El valor final de lambda después de la multiplicación es:", lambd)
    else:
        k = int(input("Ingrese el número de servidores: "))
    if metodo == "mmk_n":
        N = int(input("Ingrese el tamaño de la población (N): "))
        resultado = modelo_mmk_n(lambd, mu, k, N)
    else:
        resultado = modelo_mmk(lambd, mu, k)
elif metodo == "mg1":
    sigma = float(input("Ingrese un valor de sigma: "))
    resultado = modelo_mg1(lambd, mu, sigma)
elif metodo == "md1":
    resultado = modelo_md1(lambd, mu)
elif metodo == "mm1_q":
    Q = int(input("Ingrese el tamaño máximo de la cola (Q): "))
    resultado = modelo_mm1_q(lambd, mu, Q)
elif metodo == "mm1_n":
    N = int(input("Ingrese el tamaño de la población (N): "))
    resultado = modelo_mm1_n(lambd, mu, N)
else:
    resultado = modelo_mm1(lambd, mu)

print("Resultados:")
print("a. La probabilidad del sistema vacío:", resultado[0])
print("b. Número promedio de entidades en la cola:", resultado[1])
print("c. Número promedio de entidades en el sistema:", resultado[2])
print("d. Tiempo promedio de espera de una entidad en la cola:", resultado[3])
print("e. Tiempo promedio de una entidad en el sistema:", resultado[4])
print("f. Probailidad que el sistema esté ocupado:", resultado[5])
if metodo == "mmk" or metodo == "mm1_q" or metodo == "mm1_n":
    print("g. Probabilidad que haya 3 unidades en el sistema:")
    probabilidad_n = calcular_probabilidad_n(lambd, mu, k, 3)
    print(f"   {probabilidad_n:.4f}")
    print("h. Probabilidad de que el número de entidades en el sistema exceda a 4:")
    probabilidad_excede_4 = P0 * (lambd/mu)*5 / (math.factorial(k) * k*(5-k))
    print(f"   {probabilidad_excede_4:.4f}")
elif metodo == "mg1":
    print("g. Probabilidad que haya 3 unidades en el sistema:")
    probabilidad_n = calcular_probabilidad_n(lambd, mu, 1, 3)
    print(f"   {probabilidad_n:.4f}")
    print("g. Probabilidad que una unidad que llega tenga que esperar servivio:")
    probabilidad_espera_servicio = resultado[1] / lambd
    print(f"   {probabilidad_espera_servicio:.4f}")
elif metodo == "md1":
    print("g. Probabilidad que haya 3 unidades en el sistema:")
    probabilidad_n = calcular_probabilidad_n(lambd, mu, 1, 3)
    print(f"   {probabilidad_n:.4f}")
elif metodo == "mm1":
    print(f"g. Probabilidad que haya 3 unidades en el sistema: {(1 - resultado[0]) * (lambd/mu)**3:.4f}")
    print(f"h. Probabilidad de que el número de entidades en el sistema exceda a 4: {(lambd/mu)**5:.4f}")
```
