```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la ecuación diferencial y la condición inicial
def f(t, y):
    return 1 + y/t + (y/t)**2

# Método de Runge-Kutta de orden 4
def runge_kutta(f, t0, y0, h, num_steps):
    t_values = [t0 + i * h for i in range(num_steps + 1)]
    y_values = [y0]

    for i in range(num_steps):
        t = t_values[i]
        y = y_values[-1]

        k1 = h * f(t, y)
        k2 = h * f(t + h/2, y + k1/2)
        k3 = h * f(t + h/2, y + k2/2)
        k4 = h * f(t + h, y + k3)

        next_y = y + (k1 + 2*k2 + 2*k3 + k4)/6
        y_values.append(next_y)

    return t_values, y_values

# Parámetros del problema
t0 = 1
y0 = 0
h = 0.05
num_steps = int((3 - t0) / h)

# Obtener la solución numérica
t_values, y_values_numerical = runge_kutta(f, t0, y0, h, num_steps)

# Solución exacta
def exact_solution(t):
    return t * np.tan(np.log(t))

y_values_exact = [exact_solution(t) for t in t_values]

# Graficar la solución numérica y la solución exacta
plt.plot(t_values, y_values_numerical, label='Aproximación Numérica')
plt.plot(t_values, y_values_exact, label='Solución Exacta', linestyle='dashed')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Aproximación Numérica vs Solución Exacta')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el sistema de ecuaciones diferenciales
def f(t, y):
    x1, x2 = y
    dx1dt = x1 * (4 - 0.0003 * x1 - 0.0004 * x2)
    dx2dt = x2 * (2 - 0.0002 * x1 - 0.0004 * x1)
    return [dx1dt, dx2dt]

# Método de Runge-Kutta de orden 4 para sistemas de EDOs
def runge_kutta_system(f, t0, y0, h, num_steps):
    t_values = [t0 + i * h for i in range(num_steps + 1)]
    y_values = [y0]

    for i in range(num_steps):
        t = t_values[i]
        y = y_values[-1]

        k1 = np.multiply(h, f(t, y))
        k2 = np.multiply(h, f(t + 0.5 * h, np.add(y, 0.5 * k1)))
        k3 = np.multiply(h, f(t + 0.5 * h, np.add(y, 0.5 * k2)))
        k4 = np.multiply(h, f(t + h, np.add(y, k3)))

        next_y = np.add(y, np.multiply((k1 + 2 * k2 + 2 * k3 + k4) / 6, 1))
        y_values.append(next_y)

    return t_values, np.array(y_values)

# Parámetros del problema
t0 = 0
y0 = [10000, 10000]  # Población inicial de ambas especies
h = 0.1
num_steps = int((4 - t0) / h)

# Obtener la solución numérica
t_values, y_values = runge_kutta_system(f, t0, y0, h, num_steps)

# Graficar las soluciones aproximadas
plt.plot(t_values, y_values[:, 0], label='x1(t)')
plt.plot(t_values, y_values[:, 1], label='x2(t)')
plt.xlabel('t')
plt.ylabel('Población')
plt.title('Aproximación Numérica de la Población de Dos Especies')
plt.legend()
plt.show()

# Calcular el número de individuos de las especies 1 y 2 en el tiempo t = 2
t_index_t2 = int(2 / h)
x1_t2 = y_values[t_index_t2, 0]
x2_t2 = y_values[t_index_t2, 1]

print(f'\nNúmero de individuos en el tiempo t = 2:\nEspecie 1: {x1_t2}\nEspecie 2: {x2_t2}')
```
